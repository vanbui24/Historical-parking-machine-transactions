# Merge datasets

```python
#merging dataframes with suffixes for overlap
transactions_parkmachine = transactions.merge(parkmachine, left_on="Automat", right_on="parkomat_id", suffixes=("_transactions", "_parkmachine"))
```

```python
display(transactions_parkmachine)
```

```python
#checking column names and datatypes
print(transactions_parkmachine.columns)
print(transactions_parkmachine.dtypes)
```

```python
#dropping "0" in col[beløb]
transactions_parkmachine = transactions_parkmachine[~transactions_parkmachine["Beløb"].isin(["0"])]
```

```python
#checking drop
transactions_parkmachine.head()
```

```python
#dropping na in col ["fra", "til"]
transactions_parkmachine.dropna(subset=["Fra", "Til"], inplace=True)
```

```python
#checking Na in col Fra, Til
transactions_parkmachine.isna().sum()
```

```python
#checking df dtypes
print(transactions_parkmachine.dtypes)
print(transactions_parkmachine.shape)
```

```python
#drop columns
transactions_parkmachine.drop(["Antal", "FID", "id", "parkomat_id", "type", "vejkode", "p_zone", "beboer_zone", "vejnavn", "husnr", "placering", "ejerlavkode", "opret_dato", "ret_dato"], axis=1, inplace=True)

```

```python
#checking 
print(transactions_parkmachine.shape)
transactions_parkmachine.head()
```

```python
#add longitude and latitude columns
#stilling col wkb_geometry
transactions_parkmachine["wkb_geometry"] = transactions_parkmachine["wkb_geometry"].str.strip()
```

```python
new = transactions_parkmachine["wkb_geometry"].str.split(" ", n=3, expand=True)

transactions_parkmachine["longitude"] = new[1]
transactions_parkmachine["latitude"] = new[2]
```

```python
transactions_parkmachine.head()
```

```python
#cleaning latitude and longitude cals by removeing non Digits
transactions_parkmachine["longitude"] = transactions_parkmachine["longitude"].str.replace("(", "")
transactions_parkmachine["latitude"] = transactions_parkmachine["latitude"].str.replace(")", "")
```

```python
#replace , with . in col beløb
transactions_parkmachine["Beløb"] = transactions_parkmachine["Beløb"].str.replace(",", ".")
```

```python
#changing col names
transactions_parkmachine.rename(columns={"Automat":"automat_id", 
                                         "Lokation":"name", 
                                         "Subtype":"payment_type", 
                                         "Beløb":"amount", 
                                         "Fra":"start_time",
                                        "Til":"end_time",
                                        "Zone":"zone",
                                        "wkb_geometry":"geometry"}, inplace=True)
```

```python
#changing type
"""Automat as int64
Lokation as str
Subtype as category
Beløb as float64
Fra as datetime
Til as datetime
Zone category
wkb_geometry as object
ADD column latitude as float64
ADD column longitude as float64 """

transactions_parkmachine.dtypes
```

```python
#changing data type
transactions_parkmachine["automat_id"] = transactions_parkmachine["automat_id"].astype("int64")
transactions_parkmachine["name"] = transactions_parkmachine["name"].astype("str")
transactions_parkmachine["payment_type"] = transactions_parkmachine["payment_type"].astype("str")
transactions_parkmachine["amount"] = transactions_parkmachine["amount"].astype("float64")



transactions_parkmachine["start_time"] = pd.to_datetime(transactions_parkmachine["start_time"], format="%d-%m-%Y %H:%M")
transactions_parkmachine["end_time"] = pd.to_datetime(transactions_parkmachine["end_time"], format="%d-%m-%Y %H:%M")

transactions_parkmachine["zone"] = transactions_parkmachine["zone"].astype("category")

transactions_parkmachine["longitude"] = transactions_parkmachine["longitude"].astype("float64")
transactions_parkmachine["latitude"] = transactions_parkmachine["latitude"].astype("float64")
```
